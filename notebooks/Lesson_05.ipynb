```python
#Prueba de hipotesis

#Una fabrica produce 150 unidades por hora con una desviación estandar de 8 unidades
#Existe una máquina nueva que hace 160 unidades por hora con una desviación estandar de 8 unidades
#Total de muestras n = 35
#H0: Las máquinas son iguales

z_score1 = (160 - 150) / (8/np.sqrt(35))
#Nivel de significacia lo establecemos en alfa = 0.01, para ser cautos al tomar la decisión
print("Z_score:", z_score1)
#Rechazo la H0
p_value = 2 * (1-stats.norm.cdf(abs(z_score1)))
print("Valor:", p_value)
#Altamente improbable que ambas máquinas sean iguales
```

```python
#Un automóvil a 30 millas por hora debe recorrer máximo 64 pies al frenar
#Un fabricante de automóviles diseño un sistema nuevo de frenos, que recorre una distancia de 63.5 pies al frenar con una desviación estandar de 4 pies, habiendolo probado en 64 autos
#H0: El nuevo sistema >= 65 pies
#H1: el nuevo sistema <= 65 pies
#Nivel de significacia alfa = 0.01 => z_score = 2.36 (prueba de una sola cola)

z_score2 = (63.5 - 65) / (4/np.sqrt(64))
print(z_score2)
p_value2 = (1-stats.norm.cdf(abs(z_score2)))
print("Valor:", p_value2)
#Se recomienda el nuevo sistema ya que es probable que frene menos de 65 pies
```
