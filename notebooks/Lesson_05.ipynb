```python
import pandas as pd
import numpy as np
```

```python
#Importar el dataset del año 2010
df = pd.read_csv("../data/ACCIDENTS_GU_BCN_2010.csv", encoding="latin-1")
df
```

```python
#Muestra información del dataset
df.info
```

```python
#Convertir el formato de fecha a dd-mm-yy, forzando a leerla como cadena y utilizando una función que convierta a cadena, para luego darle el formato
df["Date"] = df[u"Dia de mes"].apply(lambda x: str(x)) + '-' + df[u"Mes de any"].apply(lambda x: str(x)) + "-2010"
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")
df.sample(6)
```

```python
#Promedio de accidentes por día en el 2010
accidents = df.groupby(df["Date"]).size()
print("accidents",accidents.mean())
```
